```python
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.xlabel('X Değerleri')
plt.ylabel('Y Değerleri')
plt.title('Çizgi Grafiği')
plt.show()
```

```python
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y, label = '1. Veriler')
plt.plot(x, y2, label = '2. Veriler')
plt.xlabel('X Değerleri')
plt.ylabel('Y Değerleri')
plt.title('Birden fazla veri seti')
plt.legend()
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [1, 2, 3, 4]

plt.scatter(x, y, color = 'green', marker = 'o')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Nokta Grafik')
plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
deger = [100, 50, 200, 500, 350]

plt.bar(kategori, deger)
plt.title('Bar Grafik')
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')
plt.show()
```

```python
secenekler = ['Elma', 'Armut', 'Portakal', 'Çilek', 'Muz']
fiyat = [20, 25, 12, 30, 40]

plt.pie(fiyat, labels=secenekler, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Pasta Grafiği')
plt.show()
```
