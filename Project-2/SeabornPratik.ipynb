```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x = np.random.randn(50)
y = np.random.randn(50)

sns.scatterplot(x=x, y=y)
plt.title('Dağılım Grafiği')
plt.xlabel('X Değerleri')
plt.ylabel('Y Değerleri')
plt.show()
```

```python
np.random.seed(0)
data = np.random.randn(1000)

sns.histplot(data=data, bins=20, kde=True)
plt.title('Histogram')
plt.xlabel('Değerler')
plt.ylabel('Frekans')
plt.show()
```

```python
np.random.seed(0)
data = np.random.randn(100,4)

sns.boxplot(data=data)
plt.title('Kutu Grafiği')
plt.xlabel('Değişkenler')
plt.ylabel('Değerler')
plt.show()
```

```python
np.random.seed(0)
data = np.random.randn(50,5)

korelasyon_matrix = np.corrcoef(data, rowvar=False)
sns.heatmap(korelasyon_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasyon Matrixi')
plt.show()
```
